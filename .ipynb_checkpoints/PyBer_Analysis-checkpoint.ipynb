```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis, label = "Boston",color="Green",marker="d",linewidth=2)
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("Pyber Fare by Month")
plt.legend()
plt.grid()
plt.show()
```

```python
#Calculates the Std Dev of the Y axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis,y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#Bar chart of the same dataset from above
plt.barh(x_axis,y_axis, color="Cyan",label="=Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.xticks(np.arange(0,51,step=5.0))
#reverses the order of the y axis
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

```python
#Scatter plot of the information
plt.scatter(x_axis,y_axis, color="aqua", label="Boston",edgecolors="black" ,s=y_axis, alpha=.2, linewidth=3)
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.show()
```

```python
#pie chart practice
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
```

```python
plt.pie(y_axis,labels=x_axis,explode=explode_values,autopct="%.1f%%")
plt.show()
```
